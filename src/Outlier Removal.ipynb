```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('train.csv')
```

```python
price = 'SalePrice'
var = 'GrLivArea'

price_vs_var = pd.concat([df[price], df[var]], axis=1)
price_vs_var.plot.scatter(x=var, y=price, ylim=(0,800000))
plt.show()
```
