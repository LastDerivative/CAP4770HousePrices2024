```python
import pandas as pd
import datetime
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#data from fullpipeline (data more consistent)

# Define the array of numbers
gb_rmse_scores = np.array([20043.46678173, 23863.77020546, 21318.58258771, 28592.3551316, 23019.45820193])
# Define the duration for the Gradient Boosting evaluation
gb_duration = datetime.timedelta(minutes=0, seconds=43, microseconds=216616)

#mean for rmse and standard deviation
gb_mean_rmse = np.mean(gb_rmse_scores)
gb_sd_rmse = np.std(gb_rmse_scores)
gb_range_rmse = np.max(gb_rmse_scores) - np.min(gb_rmse_scores)

gb_logrmse = 0.13008
```

```python
#plot the rmse scores for Gradient Boosting
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(gb_rmse_scores) + 1), gb_rmse_scores, color='blue', marker='o', linestyle='-', linewidth=2, markersize=6, label='Gradient Boosting RMSE')
plt.xlabel('Fold')
plt.ylabel('RMSE')
plt.title('Tuned RMSE Scores for Gradient Boosting')
plt.grid(True)
plt.xticks(range(1, len(gb_rmse_scores) + 1))
plt.legend()

print('Gradient Boosting RMSE mean: ', gb_mean_rmse)
print('Gradient Boosting RMSE standard deviation: ', gb_sd_rmse)
print('Gradient Boosting RMSE range: ', gb_range_rmse)
```

```python
#data from running pipeline (data more consistent)

# Define the array of numbers
nn_rmse_scores = np.array([25751.74855413, 22174.6998479, 22173.75438609, 27873.01484767, 24598.37333298])
#nn duration
nn_duration = datetime.timedelta(hours=0, minutes=10, seconds=56, microseconds=282130)

#mean for rmse, standard deviation, and range
nn_mean_rmse = np.mean(nn_rmse_scores)
nn_sd_rmse = np.std(nn_rmse_scores)
nn_range_rmse = np.max(nn_rmse_scores) - np.min(nn_rmse_scores)

nn_logrmse = 0.15049
```

```python
#plot the rmse scores for NN   
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(nn_rmse_scores) + 1), nn_rmse_scores, color='red', marker='o', linestyle='-', linewidth=2, markersize=6, label='Neural Network RMSE')
plt.xlabel('Fold')
plt.ylabel('RMSE')
plt.title('Tuned RMSE Scores for Neural Network')
plt.grid(True)
plt.xticks(range(1, len(nn_rmse_scores) + 1))
plt.legend()

print('Neural Network RMSE mean: ', nn_mean_rmse)
print('Neural Network RMSE standard deviation: ', nn_sd_rmse)
print('Neural Network RMSE range: ', nn_range_rmse)
```

```python
#data from running pipeline (data more consistent)
stacked_rmse_scores = np.array([21393.8165989, 21879.85806529, 20683.99232721, 26163.71675458, 21415.55703462])
stacked_duration = datetime.timedelta(hours=0, minutes=56, seconds=7, microseconds=341114)

#mean for rmse, standard deviation, and range
stacked_mean_rmse = np.mean(stacked_rmse_scores)
stacked_sd_rmse = np.std(stacked_rmse_scores)
stacked_range_rmse = np.max(stacked_rmse_scores) - np.min(stacked_rmse_scores)

stacked_logrmse = 0.12942
```

```python
#plot the rmse scores for Stacked
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(stacked_rmse_scores) + 1), stacked_rmse_scores, color='green', marker='o', linestyle='-', linewidth=2, markersize=6, label='Stacked RMSE')
plt.xlabel('Fold')
plt.ylabel('RMSE')
plt.title('Tuned RMSE Scores for Stacked')
plt.grid(True)
plt.xticks(range(1, len(stacked_rmse_scores) + 1))
plt.legend()

print('Stacked RMSE mean: ', stacked_mean_rmse)
print('Stacked RMSE standard deviation: ', stacked_sd_rmse)
print('Stacked RMSE range: ', stacked_range_rmse)
```

```python
#compare duration of each model
durations = {
    'Gradient Boosting': gb_duration,
    'Neural Network': nn_duration,
    'Stacked Model': stacked_duration
}

# Convert durations to minutes
durations_in_minutes = [duration.total_seconds() / 60 for duration in durations.values()]

# Plot the durations with labels
plt.figure(figsize=(10, 6))
bars = plt.bar(durations.keys(), durations_in_minutes, color=['blue', 'red', 'green'])

# Add labels and title
plt.ylabel('Duration (minutes)')
plt.title('Model Evaluation Execution Time')
plt.grid(True)

# Add a legend
plt.legend(bars, ['Gradient Boosting', 'Neural Network', 'Stacked Model'])

# Display the duration values on top of each bar
for i, duration in enumerate(durations_in_minutes):
    plt.text(i, duration + 5, f'{duration:.2f} min', ha='center', va='bottom')
```

```python
#plot the standard deviation of tuned models against each other
# Model names and corresponding standard deviation values
model_names = [
    f'Gradient Boosting\nSD: {gb_sd_rmse:.2f}',
    f'Neural Network\nSD: {nn_sd_rmse:.2f}',
    f'Stacked Model\nSD: {stacked_sd_rmse:.2f}'
]
sd_values = [gb_sd_rmse, nn_sd_rmse, stacked_sd_rmse]

# Plotting the bar chart
plt.figure(figsize=(12, 6))
bars = plt.bar(model_names, sd_values, color=['blue', 'red', 'green'])

# Adding labels and title
plt.ylabel('Standard Deviation of RMSE')
plt.title('Standard Deviation of RMSE Scores for Models')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

plt.show()

```

```python
#Compare to original rmse scores

og_rmse_scores = og_rmse_scores = np.array([32172.9301224, 33462.16289029, 61450.92029345, 31284.91229929, 24833.76276855])

#mean for rmse, standard deviation, and range
og_mean_rmse = np.mean(og_rmse_scores)
og_sd_rmse = np.std(og_rmse_scores)
og_range_rmse = np.max(og_rmse_scores) - np.min(og_rmse_scores)

og_logrmse = 0.43053
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(og_rmse_scores) + 1), og_rmse_scores, color='black', marker='o', linestyle='-', linewidth=2, markersize=6, label='Original RMSE')
plt.plot(range(1, len(gb_rmse_scores) + 1), gb_rmse_scores, color='blue', marker='o', linestyle='-', linewidth=2, markersize=6, label='Gradient Boosting RMSE')
plt.plot(range(1, len(nn_rmse_scores) + 1), nn_rmse_scores, color='red', marker='o', linestyle='-', linewidth=2, markersize=6, label='Neural Network RMSE')
plt.plot(range(1, len(stacked_rmse_scores) + 1), stacked_rmse_scores, color='green', marker='o', linestyle='-', linewidth=2, markersize=6, label='Stacked RMSE')
plt.xlabel('Fold')
plt.ylabel('RMSE')
plt.title('Tuned RMSE Scores for Each Fold Compared to Original NN model')
plt.grid(True)
plt.xticks(range(1, len(gb_rmse_scores) + 1))
plt.legend()

print('Original RMSE mean: ', og_mean_rmse)
print('Original RMSE standard deviation: ', og_sd_rmse)
print('Original RMSE range: ', og_range_rmse)
```

```python
#plotting average rmses and logrmses for each model against the original model

# Data for the chart
model_names = ['Gradient Boosting', 'Neural Network', 'Stacked Model', 'Original']
average_rmses = [gb_mean_rmse, nn_mean_rmse, stacked_mean_rmse, og_mean_rmse]
average_logrmses = [gb_logrmse, nn_logrmse, stacked_logrmse, og_logrmse]

# Plotting the Bar Chart for Average RMSE
plt.figure(figsize=(10, 6))
plt.bar(model_names, average_rmses, color=['blue', 'green', 'red', 'black'])
plt.xlabel('Model')
plt.ylabel('Average RMSE')
plt.title('Average RMSE and Log RMSE Comparison Across Models')
plt.ylim(0, max(average_rmses) + 5000)  # Adding some space at the top for better visualization

# Annotate RMSE values and Log RMSE values on top of each bar
for i, (rmse, logrmse) in enumerate(zip(average_rmses, average_logrmses)):
    plt.text(i, rmse + 500, f'{rmse:.2f}', ha='center', va='bottom', fontsize=10, color='black')
    plt.text(i, rmse + 2500, f'Log RMSE: {logrmse:.3f}', ha='center', va='bottom', fontsize=10, color='gray')

plt.show()
```
